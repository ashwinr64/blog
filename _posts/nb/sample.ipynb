## What is the corellation coefficient?
The correlation coefficient is a statistical measure of the strength of a linear relationship between two variables. Its values can range from -1 to 1. A correlation coefficient of -1 describes a perfect negative, or inverse, correlation, with values in one series rising as those in the other decline, and vice versa. A coefficient of 1 shows a perfect positive correlation, or a direct relationship. The closer to 0 the correlation coefficient is, the weaker the relationship between the variables.


$$
r = \frac{\text{cov}(X, Y)}{\sigma_X \times \sigma_Y}
$$

where,
* $cov$ is the covariance between $X$ and $Y$
* $\sigma$ represents the standard deviation.

Now, let's create two variables X and Y and plot it:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100)
Y = np.random.rand(100)


plt.scatter(X,Y)
plt.xlabel('X Value')
plt.ylabel('Y Value')

print('NumPy Correlation Method: ' + str(np.corrcoef(X, Y)[0, 1]))
```
